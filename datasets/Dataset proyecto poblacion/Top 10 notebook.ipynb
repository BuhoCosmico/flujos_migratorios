```python
import pandas as pd
import numpy as np
import json
```

```python



def leer(ruta):
    try:
        with open(ruta) as f:
            file_content = f.read()
            try:
                # Attempt to load the file as JSON
                json_data = json.loads(file_content)
                return pd.read_json(json.dumps(json_data))
            except json.decoder.JSONDecodeError:
                # If the file is not JSON, attempt to load it as CSV
                return pd.read_csv(ruta)
    except FileNotFoundError:
        print("Error: File not found.")
        return None
```
