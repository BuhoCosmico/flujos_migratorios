```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

        Se realizar치 un analisis mas profundo en el dataset de indicadores de educaci칩n para asi evaluar las diferentes series dentro de este csv y a su vez evaluar los flujos migratorios.

```python
 #Abrimos nuestro dataset que contiene los diferentes indicadores de educaci칩n
datos_educacion = pd.read_csv('indicador_educacion_normalizado.csv') 
```

```python
#chequeamos como viene la data
datos_educacion.head()
```

```python
#importamos el dataset que contiene el stock de origen de los emigrantes 
emigrantes = pd.read_csv('L-Total_emigrantes_ORIGEN_normal.csv')
```

```python
#Importamos el dataset que contiene el stock de inmigrantes y su destino
inmigrantes = pd.read_csv('L-Total_inmigrantes_DESTINO_normal.csv')
```

Del dataset de indicadores de educacion dropeamos las columnas que no son necesarias para el analisis

```python
datos_educacion.drop(columns={'Country Code'}, inplace=True)
```

```python
datos_educacion.drop(columns={'cod'},inplace=True)
```

```python
datos_educacion.head()
```

Hacemos un pivoteo en nuestro dataframe de modo que todos los indicadores queden por columna y los a침os en una sola columna

```python
data_transpuesta = (
    datos_educacion.set_index(["Country Name", "Series"])
    .rename_axis("year", axis=1)
    .stack()
    .unstack(1)
    .sort_values(['Country Name','year'])
    .reset_index()
    .rename_axis(None, axis=1)
    .rename_axis("id")
)
```

```python
data_transpuesta['All staff compensation as % of total expenditure in lower secondary public institutions (%)'].unique()
```

```python
data_transpuesta['Adult illiterate population, 15+ years, both sexes (number)'].unique()
```

```python
##Vizualizando los diferentes indicadores de educaci칩n 
for item in data_transpuesta.columns:
    print(item,end='___')
```

```python
#haciendo un filtro por los distintos a침os en el dataframe para evaluar a침o a a침o los diferentes indicadores de educacion por pais
filtro1990 = data_transpuesta[(data_transpuesta['year'] == '1990')] 
filtro1995 = data_transpuesta[(data_transpuesta['year'] == '1995')]
filtro2000 = data_transpuesta[(data_transpuesta['year'] == '2000')]
filtro2005 = data_transpuesta[(data_transpuesta['year'] == '2005')]
filtro2010 = data_transpuesta[(data_transpuesta['year'] == '2010')]
filtro2015 = data_transpuesta[(data_transpuesta['year'] == '2015')]
filtro2020 = data_transpuesta[(data_transpuesta['year'] == '2020')]
```

                        Filtrando por porcentaje de la poblaci칩n mayor de 15 a침os sin educaci칩n

```python
total_1990 = (
    filtro1990
    .groupby('Country Name')['Barro-Lee: Percentage of population age 15+ with no education']
    .mean()
    .reset_index()
    .sort_values(by='Barro-Lee: Percentage of population age 15+ with no education',ascending=False)
)
```

```python
total_1990.head()   #top 10 de los paises con poblacion mayor con mas de 15 a침os que no tienen educacion en el a침o 1990
```

```python
total_2010 = (
    filtro2010
    .groupby('Country Name')['Barro-Lee: Percentage of population age 15+ with no education']
    .mean()
    .reset_index()
    .sort_values(by='Barro-Lee: Percentage of population age 15+ with no education',ascending=False)
)
```

```python
total_2010.head()
```

se puede apreciar como este indice de personas sin educaci칩n ha venido dismuyendo en el caso de Yemen desaparece de nuestro top 5 desde 1990 hasta el 2010 veamos los flujos de migrantes para esta region 

```python
'yemen' in emigrantes['Region, development group, country or area'].unique()
```

```python
filtro_yemen = emigrantes[(emigrantes['Region, development group, country or area'] == 'yemen')]
```

```python
filtro_yemen 
```

```python
fig, ax = plt.subplots(figsize=(5,5))

yemen = filtro_yemen.loc[filtro_yemen['Region, development group, country or area'] == 'yemen', '1990':'2020']
ax.plot(yemen.values[0], color='red')
ax.set_title('Cambio en la cantidad de migrantes de Yemen durante los ultimos a침os')
ax.set_xticklabels(range(1985, 2021, 5), rotation=90)
plt.show()
```

Como era de esperarse el flujo de migrantes del la rep칰blica de Yemen ha venido en un gran aumento y puede ser una de las razones por las que cada vez hay menos cantidad de personas con mas de 15 a침os sin educaci칩n ya que hay cada vez menos datos para este indicador, una de las razones de estos flujos migratorios, para dicha region se debe a que este pais a estado involucrado en la guerra y en la actualidad se encuentra con una de las peores crisis humanitarias en el mundo
