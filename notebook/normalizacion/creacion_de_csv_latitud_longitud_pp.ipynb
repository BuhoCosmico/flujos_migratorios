```python
#Importamos las librerias de python con las que se trabajara
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from time import sleep
```

```python
#importamos de la libreria geopy para obtener las latitudes y longitudes de las diferentes regiones del dataframe
from geopy.geocoders import Nominatim
```

```python
#Extraemos el dataset que contenga todos los paises para luego asignarles la latitud y longitud
data_destinos = pd.read_csv('../../datasets/processed/proyecto_poblacion/L-destino.csv')
```

```python
#Vemos si existen duplicados
data_destinos[data_destinos.duplicated()]
```

```python
#Chequeamos que no existan valores nulos
data_destinos.isna().sum()
```

```python
#Hacemos un filtro del dataframe, ya que solo necesito la columna que contiene las diferentes regiones
filtro_pais = data_destinos['Region, development group, country or area']
```

```python
#Transformamos a lista 
filtro_pais = list(filtro_pais)
```

```python
#Creamos una lista pais para alojar todos los valores del filtro quitandole antes los espacios iniciales y finales a cada valor de la lista
lista_pais = []
for item in filtro_pais:
    lista_pais = lista_pais + [item.strip()]
```

```python
#visualizo la lista
lista_pais
```

```python
#Creo un nuevo dataframe de paises
data_paise = pd.DataFrame({'pais':lista_pais})
```

```python
data_paise
```

```python
#importamos la libreria re de python y creo una funci√≥n para eliminar los caracteres especiales dentro de los paises 
import re

def remove_special_characters(text):
    return re.sub(r"[^a-zA-Z0-9\s]+", "", text)
```

```python
#Aplico la funcion creada en el punto anterior 
data_paise['pais'] = data_paise['pais'].apply(remove_special_characters)
```

```python
#vizualizo la data
data_paise
```

```python
#Creo un objecto de la clase Nominatim, para luego crear una funci√≥n que me retorne las latitudes y longuitudes por region 
geolocator = Nominatim(user_agent="geoapiExercises")

def get_latitud_longitud(nombre_pais):
    location = geolocator.geocode(nombre_pais)
    if location is not None:
        return (location.latitude, location.longitude)
    else:
        location.latitude, location.longitude = np.NaN
        return location.latitude, location.longitude
```

```python
#Creo dos listas vacias una para guardar la latitud y otra para guardar la longitud 
lat = []
log = []
for item in data_paise['pais']:
    direccion = geolocator.geocode(item, timeout=10)
    if direccion is None:
        lat.append(None)
        log.append(None)
    else:
        latitude = direccion.latitude
        longitude = direccion.longitude
    lat.append(latitude)
    log.append(longitude)
```

Las listas lat y log seran agragadas a mi dataframe de paises

```python
#Agrego una nueva columna a mi dataframe de paises que contiene las latitudes 
data_paise['latitude'] = pd.Series(lat)
```

```python
#Agrego una nueva columna a mi dataframe de paises que contiene las longitudes
data_paise['longitude'] = pd.Series(log)
```

```python
#visualizo como quedo el dataframe
data_paise
```

```python
#creamos un nuevo csv de paises con sus latitudes y longitudes
data_destinos.to_csv('../../datasets/processed/proyecto_poblacion/L-destino_lat_long.csv', index=False)
```
